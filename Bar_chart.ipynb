```python
import pandas as pd
import plotly.graph_objects as px
```

```python
data = pd.read_csv("spotify_data.csv")
```

```python
data = data.dropna()
```

```python
columns = list(data.columns)
```

```python
data.head(2)
```

```python
for ele in range(1, len(columns)):
    plot = px.bar(data, x=columns[0], y=columns[ele], barmode = 'group')
    plot.show()
```
